```python
# resample을 사용해서 시간 간격에 따라 행 그룹핑

import pandas as pd
import numpy as np

time_index = pd.date_range('06/06/2017', periods=100000, freq='30S')

dataframe = pd.DataFrame(index=time_index)

dataframe['Sale_Amount'] = np.random.randint(1,10,100000)
dataframe
```

```python
dataframe.resample('W').sum() # 'W' 는 1주일 단위로 그룹핑 후 평균 계산
```

```python
dataframe.resample('2W').mean() # '2W' 는 2주일 단위로 그룹핑 후 평균 계산
```

```python
dataframe.resample('1M').mean()  # 1달 단위로 그룹핑 후 평균계산
```

```python
# 최대-최소 변환과정으로 정규화

import numpy as np
from sklearn import preprocessing

feature = np.array([[-500.5],
                    [-100.1],
                    [0],
                    [100.1],
                    [900.9]])

minmax_scale = preprocessing.MinMaxScaler(feature_range=(0,1))

scaled_feature = minmax_scale.fit_transform(feature)

scaled_feature
```

```python
x = np.array([[-1000.5],
              [-200.2],
              [500.5],
              [600.6],
              [9000.9]])
```

```python
# 정규화 - 데이터들의 평균 0, 표준편차 1로 변환

scaler = preprocessing.StandardScaler()

standardized = scaler.fit_transform(x)

standardized
```

```python
print(round(standardized.mean()))
print(standardized.std())
```

```python
# 결측치가 많을 경우 중간값, 사분위값 등으로 대체하는 RobustScaler

robust_scaler = preprocessing.RobustScaler()
robust_scaler.fit_transform(x)
```

```python
import numpy as np
from sklearn.covariance import EllipticEnvelope
from sklearn.datasets import make_blobs

features, _ = make_blobs(n_samples = 10,
                        n_features = 2,
                        centers = 1,
                        random_state= 1)

features[0,0] = 10000
features[1,0] = 10000

outlier_detector = EllipticEnvelope(contamination=.1)

outlier_detector.fit(features)
outlier_detector.predict(features)
```
